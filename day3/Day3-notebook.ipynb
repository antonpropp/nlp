# Привели к snake_case

```python
df.columns = (df.columns.str.strip().str.lower().str.replace(r"[^a-z0-9]+", "_", regex=True))
df.rename(columns = {'passengerid': 'passenger_id','pclass': 'p_class'}, inplace = True)
```

```python
df.head()
```

# Создаем два дф

```python
passengers = df[["passenger_id","sex","age","p_class","survived"]].copy()
```

```python
passengers.head()
```

```python
tickets = df[["passenger_id","ticket","fare","embarked"]].copy() #copy нужен чтобы не работать с view
```

```python
tickets.head()
```

```python
full_df = pd.merge(passengers,tickets, on='passenger_id', how='left') 
```

left/right выбираем ту таблицу, которую наиболее важно сохранить

```python
full_df
```

```python
print(full_df.shape) 
```

```python
agreg = full_df.groupby(['p_class','sex'])
```

groupby дает результат в сериес формате, нужно unstack делать

```python
agreg['survived'].mean().unstack()
```

```python
table = full_df.pivot_table(
    values="fare",         # какую колонку агрегировать
    index="embarked",      # что будет по строкам
    columns="p_class",     # что будет по столбцам
    aggfunc="median",      # какая агрегация (среднее, медиана, сумма и т.д.)
    fill_value=0           # чем заменить пропущенные значения
)
```

```python
table
```

```python
table.plot(kind="bar")
```

```python
table.to_parquet("/Users/podsechka/nlp-roadmap/day3/agg_day03.parquet", engine="pyarrow")
```
